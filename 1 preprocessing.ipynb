```python
import os
import cv2
import rawpy
import imageio
import numpy as np
import pandas as pd
```

```python
roots = [os.path.join('source', p) for p in os.listdir('source')]
```

```python
details = pd.read_csv('mst-e_image_details.csv')
```

```python
image_paths = []
for root in roots:
    image_paths.extend([os.path.join(root, path) for path in os.listdir(root) if path.split('.')[-1] != 'mp4'])
```

```python
img_id = details['image_ID'].tolist()
MST = details['MST'].tolist()
```

```python
len(image_paths), len(img_id), len(MST)
```

```python
labels=[]
selected_path = []
for idx in range(len(image_paths)):
    for idy in range(len(img_id)):
        id = image_paths[idx].split('\\')[-1]
        if(id == img_id[idy]):
            selected_path.append(image_paths[idx])
            labels.append(MST[idy])
```

```python
cv2.imread(selected_path[1000]).shape[1] / cv2.imread(selected_path[1000]).shape[0]
```

```python
img_shape = (int(cv2.imread(selected_path[1000]).shape[1] * 0.15), int(cv2.imread(selected_path[1000]).shape[0] * 0.15))
img_shape
```

```python
data = []
for path in selected_path:
    if(path.split('.')[-1]) != 'jpg':
        with rawpy.imread(path) as raw:
            img = raw.postprocess()
    else:
        img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, img_shape)
    data.append(img)
```

```python
print(f'Length of data: {len(data)}\nLength of Labels : {len(labels)}\nLength of path : {len(selected_path)}')
```

```python
image = np.array(data)
label = np.array(labels)
label = label.reshape(-1,1)
image.shape, label.shape
```

```python
np.save('data/data.npy', image)
np.save('data/label.npy', label)
```
