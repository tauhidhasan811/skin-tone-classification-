```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
images = np.load('data/data.npy')
labels = np.load('data/label.npy')
```

```python
count = 0
idxs = [random.randint(0,1510) for i in range(30)]
plt.figure(figsize=(30, 10))
for idx in idxs:
    plt.subplot(3, 10, count+1)
    plt.imshow(images[idx])
    plt.xticks([])
    plt.yticks([])
    plt.xlabel(labels[idx])
    count +=1
plt.show()
```
